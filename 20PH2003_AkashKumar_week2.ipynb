```python
# part b

def l_map(A, x_0, n):
    x=np.zeros(n)
    x[0]=x_0
    for i in range(n-1):
        x[i+1]=A*x[i]*(1-x[i])
    return(x)

A=3.0
x=0.93
x_1=0.94

z=[]
b=[]

for n in range(1,100):
    b.append(n)
    z=np.abs((l_map(A,x_1,n)-l_map(A,x,n))/0.01)
    
plt.plot(b,z)
#plt.yscale('log')
#plt.xscale('log')
```
